```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\Admin\\Documents\\Linear regression\\homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.Area,df.Price,color='Green',marker='*')
plt.show()
```

```python
new_df = df.drop('Price',axis='columns')
new_df
```

```python
price = df.Price
price
```

```python
# Create Linear regression object
reg = linear_model.LinearRegression()
```

```python
reg.fit(new_df,price)
```

### Predict price of with area = 3300 sqr ft


```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
