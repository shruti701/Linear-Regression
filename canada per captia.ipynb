```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:\\pandas\\canada_per_capita_income.csv')
df
```

```python
df.rename(columns={'per capita income (US$)':'per_capita_income'}, inplace=True)
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('Per capitia income')
plt.scatter(df.year,df.per_capita_income,color='Green',marker='*')
plt.show()
```

```python
new_df = df.drop('per_capita_income',axis='columns')
new_df
```

```python
per_capita_income=df.per_capita_income
per_capita_income
```

```python
# Create liner regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,per_capita_income)
```

```python
reg.predict([[2016]])
```
